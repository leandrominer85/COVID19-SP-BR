```python
#Importação dos modulos
import pandas as pd
import re
import plotly.graph_objects as go
import plotly.express as px
```

```python
#Importação do banco de dados e padronização
df = pd.read_csv(r"D:\Dropbox\Pessoal\Python\COVID\tabelas_analise\infectados_municipio.csv")
df.rename(columns={"Unnamed: 0": "municipio"}, inplace=True)
df.set_index('municipio', inplace=True)
df.fillna(0,inplace=True)
df.head()
```

```python
#Estilização do cabeçalho


dias=['26_mar', '27_mar', '28_mar', '29_mar', '30_mar',
       '01_abr', '02_abr', '03_abr', '04_abr', '05_abr', '06_abr', '07_abr',
       '08_abr', '09_abr', '10_abr', '11_abr', '12_abr', '13_abr', '14_abr',
       '15_abr', '16_abr', '17_abr', '18_abr', '19_abr', '20_abr', '21_abr',
       '22_abr', '23_abr', '24_abr', '25_abr', '26_abr', '27_abr', '28_abr',
       '29_abr', '30_abr', '01_mai', '02_mai', '03_mai', '04_mai', '05_mai',
       '06_mai', '07_mai', '08_mai', '09_mai', '10_mai', '11_mai', '12_mai',
       '13_mai', '14_mai', '15_mai', '16_mai', '17_mai','18_mai', '19_mai', '20_mai',
       '21_mai', '22_mai', '23_mai', '24-mai', '25-mai']


dias = [re.sub("_mar", '/03', x) for x in dias]
dias = [re.sub("_abr", '/04', x) for x in dias]
dias = [re.sub("_mai", '/05', x) for x in dias]
dias = [re.sub("-mai", '/05', x) for x in dias]



df.rename(columns=dict(zip(df.loc[:, '26_mar':'25-mai'], dias)),inplace=True)
df.iloc[:,1:53].columns = pd.to_datetime(df.iloc[:,1:53].columns, dayfirst=True, errors= "coerce")




```

```python
#Criando o dataframe transposto
df_t=df.T
df_t.head()
```

```python
#Inicialização da figura

#a partir do codigo disponivel em : https://stackoverflow.com/questions/59406167/plotly-how-to-filter-a-pandas-dataframe-using-a-dropdown-menu

fig=go.Figure()

# Scatter com uma linha
fig.add_trace(go.Scatter(x=df_t.index,
                             y=df_t[df_t.columns[0]],
                             visible=True)
             )


fig.update_layout(title="Casos Covid por município no Estado de São Paulo",yaxis_title="Casos",xaxis_title="Dias")



updatemenu=[]
buttons=[]

# Criação do botão de dropdown a partir de um dicionário com os argumentos da iteração do df
for col in df_t.columns:

    buttons.append(dict(method='restyle',
                        label=col,
                        visible=True,
                        args=[{'y':[df_t[col]],
                               'x':[df_t.index],
                               'type':'scatter'}, [0]],
                        )
                  )


updatemenu=[]
your_menu=dict()
updatemenu.append(your_menu)

updatemenu[0]['buttons']=buttons
updatemenu[0]['direction']='down'
updatemenu[0]['showactive']=True

# Adicionando o menu dropdown

fig.update_layout(showlegend=False, updatemenus=updatemenu)
fig.show()
```

```python
fig.write_html("grafico.html")
```
